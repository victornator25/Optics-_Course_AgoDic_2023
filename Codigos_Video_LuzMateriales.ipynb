#Coeficiente de propagación y coeficiente de propagación para dieléctricos con pérdidas

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sc

def a(w, s, mu, e):
    return w * np.sqrt((mu * e / 2) * (np.sqrt(1 + (s / (w * e)) ** 2) - 1))

def b(w, s, mu, e):
    return w * np.sqrt(mu * e / 2 * (np.sqrt(1 + (s / (w * e)) ** 2) + 1))

# Onda en dieléctrico con pérdidas.
s = 1e-12
er = 7.8
mu = 1
e = sc.epsilon_0 * er

w = np.linspace(1e15, 10e15, 10000)  # Rango de valores para w (frecuencias en Hz)

plt.plot(w, a(w, s, mu, e), color='pink', label="$\\alpha(w)$")
plt.plot(w, b(w, s, mu, e), color='purple', label="$\\beta(w)$")
plt.xlabel("$\\omega$ [Hz]")
plt.title("$\\alpha$ y $\\beta$ en Dieléctricos con Pérdidas ")
plt.legend()
plt.show()
```

#Coef. de propagación y atenuación para dieléctricos sin pérdidas

```python
# Parámetros constantes
er = 7.8
mu = 1
e = sc.epsilon_0 * er

# Rango de frecuencias
w = np.linspace(1e15, 5e15, 1000)  # Frecuencias en Hz

# Cálculo de beta
beta = w * np.sqrt(e * mu)

plt.plot(w, beta, color='lightpink', label="$\\beta(\\omega) = \\omega\\sqrt{\\varepsilon\\mu}$")
plt.xlabel("$\\omega$ [Hz]")
plt.title("α y β en Dieléctricos sin Pérdidas  σ << wε")
plt.legend()
plt.show()
```

Para el vacío observamos el mismo comportamiento que para un dieléctrico sin pérdidas

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sc

# Parámetros constantes
er = 7.8
mu = 1
e = sc.epsilon_0 * er

# Rango de frecuencias
w = np.linspace(1e15, 5e15, 1000)  # Frecuencias en Hz

# Cálculo de beta
beta = w * np.sqrt(e * mu)

plt.plot(w, beta, color='navy', label="$\\beta(\\omega) = \\omega\\sqrt{\\varepsilon\\mu}$")
plt.xlabel("$\\omega$ [Hz]")
plt.title("α y β en Vacio σ=0")
plt.legend()
plt.show()
```

#Para conductores

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sc

# Parámetros constantes
er = 7.8
mu = 1
sigma = 1e-12  # Conductividad
e = sc.epsilon_0 * er

# Rango de frecuencias
w = np.linspace(1e15, 5e15, 1000)  # Frecuencias en Hz

# Cálculo de alpha y beta
alpha = beta = w * e * mu / (2 * np.sqrt(sigma * w * e))

plt.plot(w, alpha, color='red', label="$\\alpha(\\omega) = \\beta(\\omega) = \\frac{\\omega\\varepsilon\\mu}{2\\sqrt{\\sigma\\omega\\varepsilon}}$")
plt.xlabel("$\\omega$ [Hz]")
plt.title("$\\alpha$ y $\\beta$ Buenos conductores σ >> wε")
plt.legend()
plt.show()
```
