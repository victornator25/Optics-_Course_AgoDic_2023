#Secciones cónicas

```python
import numpy as np
import matplotlib.pyplot as plt

#coeficientes:
A=2
B=4
C=2
D=7
E=-3
F=0

x_val=np.linspace(-6,6,200)
y_val=np.linspace(-6,6,200)

x,y=np.meshgrid(x_val, y_val)

#Ecuación general de segundo grado:
z=A*x**2 + B*x*y + C*y**2 + D*x + E*y + F

plt.contour(x,y,z, levels=[0], colors='b')
plt.grid(True)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('x')
plt.title('Lugar Geométrico: Parábola')

plt.gca().set_aspect('equal', adjustable='box') #para que los ejes tengan el mismo tamaño
plt.show()

```

#Gráficas de Ondas de Superposición

ondas superpuestas constructiva y destructivamente


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las ondas
t = np.linspace(0, 20, 1000)  # Valores en el tiempo
E01 = 1.0  # Amplitud de la primera onda
E02 = 0.7  # Amplitud de la segunda onda
alpha1 = np.pi/4  # Fase de la primera onda
alpha2 = np.pi/6  # Fase de la segunda onda
omega1 = 1.0  # Frecuencia angular de la primera onda
omega2 = 1.5  # Frecuencia angular de la segunda onda

# Cálculo de las ondas individuales
E1 = E01 * np.cos(alpha1 - omega1 * t)
E2 = E02 * np.cos(alpha2 - omega2 * t)

# Cálculo de las ondas resultantes
Er_constructiva = (E01 + E02) * np.cos(alpha1 - omega1 * t)
Er_destructiva = (E01 - E02) * np.cos(alpha2 - omega1 * t)

# Graficar las ondas individuales y las ondas resultantes en el mismo plot
plt.figure(figsize=(10, 6))
plt.plot(t, E1, label='Onda 1')
plt.plot(t, E2, label='Onda 2')
plt.plot(t, Er_constructiva, label='Superposición Constructiva')
plt.plot(t, Er_destructiva, label='Superposición Destructiva')
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.title('Ondas Individuales y Superposiciones')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las ondas
N = 30  # Número de ondas
x = np.linspace(0, 2*np.pi, 1000)  # Valores en el eje x
E0_i = np.random.rand(N)  # Valores aleatorios para E0i entre 0 y 1
alpha_i = np.random.uniform(0, 2*np.pi, N)  # Ángulos aleatorios para alpha_i entre 0 y 2*pi

# Cálculo de E0 para cada punto en x
E0 = np.zeros(len(x))
for i in range(N):
    E0 += E0_i[i] * np.cos(alpha_i[i] - x)

# Graficar la superposición de las ondas
plt.figure(figsize=(10, 6))
plt.plot(x, E0)
plt.xlabel('Eje x')
plt.ylabel('E0')
plt.title('Superposición de 30 Ondas')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las ondas
x = np.linspace(0, 15, 1000)  # Valores en el eje x
N1 = 30  # Número de ondas para el primer conjunto
N2 = 50  # Número de ondas para el segundo conjunto

# Generar valores aleatorios para E0i y alpha_i para ambos conjuntos de ondas
E0_i1 = np.random.rand(N1)
alpha_i1 = np.random.uniform(0, 2*np.pi, N1)
E0_i2 = np.random.rand(N2)
alpha_i2 = np.random.uniform(0, 2*np.pi, N2)

# Cálculo de E0 para cada conjunto de ondas
E0_1 = np.zeros(len(x))
E0_2 = np.zeros(len(x))

for i in range(N1):
    E0_1 += E0_i1[i] * np.cos(alpha_i1[i] - x)

for i in range(N2):
    E0_2 += E0_i2[i] * np.cos(alpha_i2[i] - x)

# Graficar ambas superposiciones de ondas en la misma imagen
plt.figure(figsize=(10, 6))
plt.plot(x, E0_1, label='Superposición de 30 Ondas')
plt.plot(x, E0_2, label='Superposición de 50 Ondas')
plt.xlabel('Eje x')
plt.ylabel('E0')
plt.title('Superposición de Ondas (30 y 50)')
plt.legend()
plt.grid(True)
plt.show()
```
