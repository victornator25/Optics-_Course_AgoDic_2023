```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk


# Índices de refracción relativos
n1 = 1.0  # Índice de refracción del medio incidente (por ejemplo, aire)
n2 = 1.5  # Índice de refracción del medio transmitido (por ejemplo, vidrio)

# Ángulos de incidencia en radianes
theta_i = np.linspace(0, np.pi/2, 100)

# Cálculo de los coeficientes de Fresnel
R_TE = (np.cos(theta_i) - np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
R_TM = (-(n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
T_TE = (2 * np.cos(theta_i)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
T_TM = (2* (n2/n1)* np.cos(theta_i)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))

# Ángulo de Brewster
theta_Brewster = np.arctan(n2/n1)

# Creación de gráficos
plt.figure(figsize=(10, 6))
plt.plot(np.degrees(theta_i), R_TE, label='R_TE',  color='pink')
plt.plot(np.degrees(theta_i), R_TM, label='R_TM', color='purple')
plt.plot(np.degrees(theta_i), T_TE, label='T_TE', linestyle='--', color='magenta')
plt.plot(np.degrees(theta_i), T_TM, label='T_TM', linestyle='--', color='slategrey')
plt.axvline(np.degrees(theta_Brewster), color='darkcyan', linestyle='--', label=f'Ángulo de Brewster ({np.degrees(theta_Brewster):.2f}°)')
plt.xlabel('Ángulo de Incidencia (grados)', color='black')
plt.ylabel('Coeficientes de Fresnel')
plt.legend()
plt.grid()
plt.title('Coeficientes de Fresnel en función del ángulo de incidencia')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Índices de refracción relativos
n1 = 1.0  # Índice de refracción del medio incidente (por ejemplo, aire)
n2 = 1.5  # Índice de refracción del medio transmitido (por ejemplo, vidrio)

# Ángulos de incidencia en radianes
theta_i = np.linspace(0, np.pi/2, 100)

# Cálculo de los coeficientes de Fresnel
R_TE = (np.cos(theta_i) - np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
R_TM = (-(n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
T_TE = (2 * np.cos(theta_i)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))
T_TM = (2* (n2/n1)* np.cos(theta_i)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2))

# Escalar los coeficientes de Fresnel
datos = 'Práctica 5 - Fresnel.xlsx'
df_1 = pd.read_excel(datos, sheet_name='Transmisión')
escala = df_1['Amplitud (Mili-Voltaje)'].max()  # Escala para coeficientes
R_TE_escalado = R_TE * escala
R_TM_escalado = R_TM * escala
T_TE_escalado = T_TE * escala
T_TM_escalado = T_TM * escala

# Normalizar los coeficientes de Fresnel al rango de 0 a 1
def normalize_to_01(arr):
    min_val = arr.min()
    max_val = arr.max()
    return (arr - min_val) / (max_val - min_val)

R_TE_escalado = normalize_to_01(R_TE_escalado)
R_TM_escalado = normalize_to_01(R_TM_escalado)
T_TE_escalado = normalize_to_01(T_TE_escalado)
T_TM_escalado = normalize_to_01(T_TM_escalado)

# Creación de gráficos
plt.figure(figsize=(10, 6))
plt.plot(np.degrees(theta_i), R_TE_escalado, label='R_TE', color='pink')
plt.plot(np.degrees(theta_i), R_TM_escalado, label='R_TM', color='purple')
plt.plot(np.degrees(theta_i), T_TE_escalado, label='T_TE', linestyle='--', color='magenta')
plt.plot(np.degrees(theta_i), T_TM_escalado, label='T_TM', linestyle='--', color='slategrey')

# Leer los datos del archivo Excel (Transmisión)
df_1 = pd.read_excel(datos, sheet_name='Transmisión')

# Normalizar amplitud de Transmisión al rango de 0 a 1
df_1['Amplitud (Mili-Voltaje)'] = normalize_to_01(df_1['Amplitud (Mili-Voltaje)'])

# Gráfico de Transmisión desde el archivo Excel
plt.plot(df_1['Grados'], df_1['Amplitud (Mili-Voltaje)'], color='navy', marker='*', linestyle='-', label='Transmisión')

# Leer los datos del archivo Excel (Reflexión)
df_2 = pd.read_excel(datos, sheet_name='Reflexión')

# Normalizar amplitud de Reflexión al rango de 0 a 1
df_2['Amplitud(Mili-Volts)'] = normalize_to_01(df_2['Amplitud(Mili-Volts)'])

# Gráfico de Reflexión
df2 = df_2.dropna()
plt.scatter(df2['Grados'], df2['Amplitud(Mili-Volts)'], color='steelblue', label='Reflexión')
plt.plot(df2['Grados'], df2['Amplitud(Mili-Volts)'], color='steelblue')

# Ángulo de Brewster
theta_Brewster = np.arctan(n2/n1)
plt.axvline(np.degrees(theta_Brewster), color='darkcyan', linestyle='--', label=f'Ángulo de Brewster ({np.degrees(theta_Brewster):.2f}°)')

# Configurar etiquetas y leyenda
plt.xlabel('Ángulo [grados]')
plt.ylabel('Intensidad')
plt.title('Transmisión y Reflexión')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Índices de refracción relativos
n1 = 1.0  # Índice de refracción del medio incidente (por ejemplo, aire)
n2 = 1.5  # Índice de refracción del medio transmitido (por ejemplo, vidrio)

# Ángulos de incidencia en radianes
theta_i = np.linspace(0, np.pi/2, 100)

# Cálculo de los coeficientes de Fresnel
R_TE = (np.cos(theta_i) - np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) * 100
R_TM = (-(n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) * 100
T_TE = (2 * np.cos(theta_i)) / (np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) * 100
T_TM = (2* (n2/n1)* np.cos(theta_i)) / ((n2/n1)**2 * np.cos(theta_i) + np.sqrt((n2/n1)**2 - np.sin(theta_i)**2)) * 100

# Normalizar los valores entre -1 y 1
R_TE = R_TE / max(abs(R_TE))
R_TM = R_TM / max(abs(R_TM))
T_TE = T_TE / max(abs(T_TE))
T_TM = T_TM / max(abs(T_TM))

# Creación de gráficos sin escalamiento
plt.figure(figsize=(10, 6))
plt.plot(np.degrees(theta_i), R_TE, label='R E', color='green')
plt.plot(np.degrees(theta_i), R_TM, label='R M', color='purple')
plt.plot(np.degrees(theta_i), T_TE, label='T E', linestyle='--', color='magenta')
plt.plot(np.degrees(theta_i), T_TM, label='T M', linestyle='--', color='slategrey')

# Leer los datos del archivo Excel (Transmisión)
datos = 'Práctica 5 - Fresnel.xlsx'
df_1 = pd.read_excel(datos, sheet_name='Transmisión')
df_1['Amplitud (Mili-Voltaje)'] = (df_1['Amplitud (Mili-Voltaje)'] / 620) * 100
df_1['Amplitud (Mili-Voltaje)'] = df_1['Amplitud (Mili-Voltaje)'] / max(abs(df_1['Amplitud (Mili-Voltaje)']))

# Gráfico de Transmisión desde el archivo Excel
plt.plot(df_1['Grados'], df_1['Amplitud (Mili-Voltaje)'], color='darkorange', marker='*', linestyle='-', label='Transmisión')

# Leer los datos del archivo Excel (Reflexión)
df_2 = pd.read_excel(datos, sheet_name='Reflexión')
df2 = df_2.dropna()
df2['Not used'] = (df2['Not used'] / 620) * 100
df2['Not used'] = df2['Not used'] / max(abs(df2['Not used']))

# Gráfico de Reflexión
plt.scatter(df2['Grados'], df2['Not used'], color='steelblue', label='Reflexión')
plt.plot(df2['Grados'], df2['Not used'], color='steelblue')

# Ángulo de Brewster
theta_Brewster = np.arctan(n2/n1)
plt.axvline(np.degrees(theta_Brewster), color='darkcyan', linestyle='--', label=f'Ángulo de Brewster ({np.degrees(theta_Brewster):.2f}°)')

# Configurar etiquetas y leyenda
plt.xlabel('Ángulo [°]')
plt.ylabel('Valor del Coeficiente (Normalizado)')
plt.title('Transmisión y Reflexión')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()
```
