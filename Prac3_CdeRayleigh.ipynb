```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def plot_gaussian_fit(x, y, A_fit, mu_fit, sigma_fit, label, y_err=20, uno_e_value=None, uno_e_c_value=None, titulo=None):
    # Encontrar las coordenadas de intersección
    x_interseccion_1_e = fsolve(lambda x: gaussiana(x, A_fit, mu_fit, sigma_fit) - uno_e_value, 0)
    x_interseccion_1_e_c = fsolve(lambda x: gaussiana(x, A_fit, mu_fit, sigma_fit) - uno_e_c_value, 0)

    # Filtrar los valores de intersección donde x > 0
    x_interseccion_1_e = [x for x in x_interseccion_1_e if x > 0]
    x_interseccion_1_e_c = [x for x in x_interseccion_1_e_c if x > 0]

    # Crear la figura y los ejes
    plt.figure(figsize=(10, 6))

    # Scatter plot de los datos con barras de error de ±y_err (en mV)
    plt.errorbar(x, y, yerr=y_err, label='Datos', fmt='bo', capsize=5)

    # Gráfico del ajuste exponencial
    x_fit = np.linspace(min(x), 40, 100)  # Rango extendido hasta 40 para el ajuste
    plt.plot(x_fit, gaussiana(x_fit, A_fit, mu_fit, sigma_fit), label='Ajuste Exponencial', color='pink')

    # Gráfica de las líneas horizontales 1/e y 1/e^2 si se proporcionan los valores
    if uno_e_value is not None:
        plt.axhline(y=uno_e_value, color='k', linestyle='--', label='$1/e$')
    if uno_e_c_value is not None:
        plt.axhline(y=uno_e_c_value, color='purple', linestyle='--', label='$1/e^2$')

    # Personalizar el gráfico
    if titulo is not None:
        plt.title(titulo)
    plt.xlabel('Pasos x 25 $ \mu m$')
    plt.ylabel(' Amplitud [V]')
    plt.legend()

    # Etiquetas de los puntos de intersección y creación de la matriz
    intersecciones = []

    for x_value in x_interseccion_1_e:
        y_value = gaussiana(x_value, A_fit, mu_fit, sigma_fit)
        intersecciones.append([x_value, y_value])
        plt.annotate(f'Intersección 1/e: ({x_value:.2f}, {y_value:.2f})',
                     xy=(x_value, y_value), xytext=(x_value + 5, y_value + 0.1),
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.3"))
    for x_value in x_interseccion_1_e_c:
        y_value = gaussiana(x_value, A_fit, mu_fit, sigma_fit)
        intersecciones.append([x_value, y_value])
        plt.annotate(f'Intersección $1/e^2$: ({x_value:.2f}, {y_value:.2f})',
                     xy=(x_value, y_value),
                     xytext=(x_value + 5, y_value + 0.1),
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.3"))

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()

    # Devolver solo los valores de intersección en x donde x > 0 y la matriz de intersecciones
    return [x for x in x_interseccion_1_e if x > 0], [x for x in x_interseccion_1_e_c if x > 0], intersecciones

# Uso de la función y obtención de los valores de intersección y la matriz
x_interseccion_1_e, x_interseccion_1_e_c, intersecciones = plot_gaussian_fit(x_1, y_1, A_fit_1, mu_fit_1, sigma_fit_1, label='Datos', y_err=20, uno_e_value=uno_e(y_1), uno_e_c_value=uno_e_c(y_1), titulo='Corte 1 - Ajuste 1')

# Imprimir la matriz de intersecciones
print("Matriz de Intersecciones:")
for i, (x_val, y_val) in enumerate(intersecciones, start=1):
    print(f"Intersección {i}: ({x_val:.2f}, {y_val:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussiana(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

def plot_gaussian_fit(x, y, A_fit, mu_fit, sigma_fit, label, y_err=20, uno_e_value=None, uno_e_c_value=None, titulo=None):
    # Crear la figura y los ejes
    plt.figure(figsize=(10, 6))

    # Scatter plot de los datos con barras de error de ±y_err (en mV)
    plt.errorbar(x, y, yerr=y_err, label='Datos',color='teal', fmt='bo', capsize=5)

    # Gráfico del ajuste exponencial
    x_fit = np.linspace(min(x), max(x), 1000)
    y_fit = gaussiana(x_fit, A_fit, mu_fit, sigma_fit)
    plt.plot(x_fit, y_fit, label='Ajuste Exponencial', color='navy')

    # Gráfica de las líneas horizontales 1/e y 1/e^2 si se proporcionan los valores
    if uno_e_value is not None:
        plt.axhline(y=uno_e_value, color='k', linestyle='--', label='$1/e$')
    if uno_e_c_value is not None:
        plt.axhline(y=uno_e_c_value, color='purple', linestyle='--', label='$1/e^2$')

    # Personalizar el gráfico
    if titulo is not None:
        plt.title(titulo)
    plt.xlabel('Pasos x 25 $ \mu m$')
    plt.ylabel('Amplitud [V]')
    plt.legend()

    # Calcular las coordenadas de intersección
    interseccion_1_e = None
    interseccion_1_e_c = None

    for i in range(len(x_fit)):
        if uno_e_value is not None and y_fit[i] <= uno_e_value:
            interseccion_1_e = (x_fit[i], y_fit[i])
            break

    for i in range(len(x_fit)):
        if uno_e_c_value is not None and y_fit[i] <= uno_e_c_value:
            interseccion_1_e_c = (x_fit[i], y_fit[i])
            break

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()

    # Devolver las coordenadas de intersección
    return interseccion_1_e, interseccion_1_e_c

# Uso de la función y obtención de las coordenadas de intersección para el primer gráfico
interseccion_1_e_1, interseccion_1_e_c_1 = plot_gaussian_fit(x_1, y_1, A_fit_1, mu_fit_1, sigma_fit_1, label='Datos', y_err=20, uno_e_value=uno_e(y_1), uno_e_c_value=uno_e_c(y_1), titulo='Corte 1 - Ajuste 1')

# Uso de la función y obtención de las coordenadas de intersección para el segundo gráfico
interseccion_1_e_2, interseccion_1_e_c_2 = plot_gaussian_fit(x_2, y_2, A_fit_2, mu_fit_2, sigma_fit_2, label='Datos', y_err=20, uno_e_value=uno_e(y_2), uno_e_c_value=uno_e_c(y_2), titulo='Corte 2 - Ajuste 2')

# Uso de la función y obtención de las coordenadas de intersección para el tercer gráfico
interseccion_1_e_3, interseccion_1_e_c_3 = plot_gaussian_fit(x_3, y_3, A_fit_3, mu_fit_3, sigma_fit_3, label='Datos', y_err=20, uno_e_value=uno_e(y_3), uno_e_c_value=uno_e_c(y_3), titulo='Corte 3 - Ajuste 3')

# Uso de la función y obtención de las coordenadas de intersección para el cuarto gráfico
interseccion_1_e_4, interseccion_1_e_c_4 = plot_gaussian_fit(x_4, y_4, A_fit_4, mu_fit_4, sigma_fit_4, label='Datos', y_err=20, uno_e_value=uno_e(y_4), uno_e_c_value=uno_e_c(y_4), titulo='Corte 4 - Ajuste 4')

# Uso de la función y obtención de las coordenadas de intersección para el quinto gráfico
interseccion_1_e_5, interseccion_1_e_c_5 = plot_gaussian_fit(x_5, y_5, A_fit_5, mu_fit_5, sigma_fit_5, label='Datos', y_err=20, uno_e_value=uno_e(y_5), uno_e_c_value=uno_e_c(y_5), titulo='Corte 5 - Ajuste 5')

# Uso de la función y obtención de las coordenadas de intersección para el sexto gráfico
interseccion_1_e_6, interseccion_1_e_c_6 = plot_gaussian_fit(x_6, y_6, A_fit_6, mu_fit_6, sigma_fit_6, label='Datos', y_err=20, uno_e_value=uno_e(y_6), uno_e_c_value=uno_e_c(y_6), titulo='Corte 6 - Ajuste 6')

# Uso de la función y obtención de las coordenadas de intersección para el séptimo gráfico
interseccion_1_e_7, interseccion_1_e_c_7 = plot_gaussian_fit(x_7, y_7, A_fit_7, mu_fit_7, sigma_fit_7, label='Datos', y_err=20, uno_e_value=uno_e(y_7), uno_e_c_value=uno_e_c(y_7), titulo='Corte 7 - Ajuste 7')

# Uso de la función y obtención de las coordenadas de intersección para el octavo gráfico
interseccion_1_e_8, interseccion_1_e_c_8 = plot_gaussian_fit(x_8, y_8, A_fit_8, mu_fit_8, sigma_fit_8, label='Datos', y_err=20, uno_e_value=uno_e(y_8), uno_e_c_value=uno_e_c(y_8), titulo='Corte 8 - Ajuste 8')

# Uso de la función y obtención de las coordenadas de intersección para el noveno gráfico
interseccion_1_e_9, interseccion_1_e_c_9 = plot_gaussian_fit(x_9, y_9, A_fit_9, mu_fit_9, sigma_fit_9, label='Datos', y_err=20, uno_e_value=uno_e(y_9), uno_e_c_value=uno_e_c(y_9), titulo='Corte 9 - Ajuste 9')

# Uso de la función y obtención de las coordenadas de intersección para el décimo gráfico
interseccion_1_e_10, interseccion_1_e_c_10 = plot_gaussian_fit(x_10, y_10, A_fit_10, mu_fit_10, sigma_fit_10, label='Datos', y_err=20, uno_e_value=uno_e(y_10), uno_e_c_value=uno_e_c(y_10), titulo='Corte 10 - Ajuste 10')

# Uso de la función y obtención de las coordenadas de intersección para el undécimo gráfico
interseccion_1_e_11, interseccion_1_e_c_11 = plot_gaussian_fit(x_11, y_11, A_fit_11, mu_fit_11, sigma_fit_11, label='Datos', y_err=20, uno_e_value=uno_e(y_11), uno_e_c_value=uno_e_c(y_11), titulo='Corte 11 - Ajuste 11')

# Uso de la función y obtención de las coordenadas de intersección para el duodécimo gráfico
interseccion_1_e_12, interseccion_1_e_c_12 = plot_gaussian_fit(x_12, y_12, A_fit_12, mu_fit_12, sigma_fit_12, label='Datos', y_err=20, uno_e_value=uno_e(y_12), uno_e_c_value=uno_e_c(y_12), titulo='Corte 12 - Ajuste 12')

# Uso de la función y obtención de las coordenadas de intersección para el decimotercer gráfico
interseccion_1_e_13, interseccion_1_e_c_13 = plot_gaussian_fit(x_13, y_13, A_fit_13, mu_fit_13, sigma_fit_13, label='Datos', y_err=20, uno_e_value=uno_e(y_13), uno_e_c_value=uno_e_c(y_13), titulo='Corte 13 - Ajuste 13')

# Uso de la función y obtención de las coordenadas de intersección para el decimocuarto gráfico
interseccion_1_e_14, interseccion_1_e_c_14 = plot_gaussian_fit(x_14, y_14, A_fit_14, mu_fit_14, sigma_fit_14, label='Datos', y_err=20, uno_e_value=uno_e(y_14), uno_e_c_value=uno_e_c(y_14), titulo='Corte 14 - Ajuste 14')

# Uso de la función y obtención de las coordenadas de intersección para el decimoquinto gráfico
interseccion_1_e_15, interseccion_1_e_c_15 = plot_gaussian_fit(x_15, y_15, A_fit_15, mu_fit_15, sigma_fit_15, label='Datos', y_err=20, uno_e_value=uno_e(y_15), uno_e_c_value=uno_e_c(y_15), titulo='Corte 15 - Ajuste 15')

# Uso de la función y obtención de las coordenadas de intersección para el decimosexto gráfico
interseccion_1_e_16, interseccion_1_e_c_16 = plot_gaussian_fit(x_16, y_16, A_fit_16, mu_fit_16, sigma_fit_16, label='Datos', y_err=20, uno_e_value=uno_e(y_16), uno_e_c_value=uno_e_c(y_16), titulo='Corte 16 - Ajuste 16')

# Uso de la función y obtención de las coordenadas de intersección para el decimoséptimo gráfico
interseccion_1_e_17, interseccion_1_e_c_17 = plot_gaussian_fit(x_17, y_17, A_fit_17, mu_fit_17, sigma_fit_17, label='Datos', y_err=20, uno_e_value=uno_e(y_17), uno_e_c_value=uno_e_c(y_17), titulo='Corte 17 - Ajuste 17')


# Imprimir las coordenadas de intersección para todos los gráficos
print("Coordenadas de Intersección - Corte 1:")
print("Intersección 1/e:", interseccion_1_e_1)
print("Intersección 1/e^2:", interseccion_1_e_c_1)

print("\nCoordenadas de Intersección - Corte 2:")
print("Intersección 1/e:", interseccion_1_e_2)
print("Intersección 1/e^2:", interseccion_1_e_c_2)

print("\nCoordenadas de Intersección - Corte 3:")
print("Intersección 1/e:", interseccion_1_e_3)
print("Intersección 1/e^2:", interseccion_1_e_c_3)

print("\nCoordenadas de Intersección - Corte 4:")
print("Intersección 1/e:", interseccion_1_e_4)
print("Intersección 1/e^2:", interseccion_1_e_c_4)

print("\nCoordenadas de Intersección - Corte 5:")
print("Intersección 1/e:", interseccion_1_e_5)
print("Intersección 1/e^2:", interseccion_1_e_c_5)

print("\nCoordenadas de Intersección - Corte 6:")
print("Intersección 1/e:", interseccion_1_e_6)
print("Intersección 1/e^2:", interseccion_1_e_c_6)

print("\nCoordenadas de Intersección - Corte 7:")
print("Intersección 1/e:", interseccion_1_e_7)
print("Intersección 1/e^2:", interseccion_1_e_c_7)

print("\nCoordenadas de Intersección - Corte 8:")
print("Intersección 1/e:", interseccion_1_e_8)
print("Intersección 1/e^2:", interseccion_1_e_c_8)

print("\nCoordenadas de Intersección - Corte 9:")
print("Intersección 1/e:", interseccion_1_e_9)
print("Intersección 1/e^2:", interseccion_1_e_c_9)

print("\nCoordenadas de Intersección - Corte 10:")
print("Intersección 1/e:", interseccion_1_e_10)
print("Intersección 1/e^2:", interseccion_1_e_c_10)

print("\nCoordenadas de Intersección - Corte 11:")
print("Intersección 1/e:", interseccion_1_e_11)
print("Intersección 1/e^2:", interseccion_1_e_c_11)

print("\nCoordenadas de Intersección - Corte 12:")
print("Intersección 1/e:", interseccion_1_e_12)
print("Intersección 1/e^2:", interseccion_1_e_c_12)

print("\nCoordenadas de Intersección - Corte 13:")
print("Intersección 1/e:", interseccion_1_e_13)
print("Intersección 1/e^2:", interseccion_1_e_c_13)

print("\nCoordenadas de Intersección - Corte 14:")
print("Intersección 1/e:", interseccion_1_e_14)
print("Intersección 1/e^2:", interseccion_1_e_c_14)

print("\nCoordenadas de Intersección - Corte 15:")
print("Intersección 1/e:", interseccion_1_e_15)
print("Intersección 1/e^2:", interseccion_1_e_c_15)

print("\nCoordenadas de Intersección - Corte 16:")
print("Intersección 1/e:", interseccion_1_e_16)
print("Intersección 1/e^2:", interseccion_1_e_c_16)

print("\nCoordenadas de Intersección - Corte 17:")
print("Intersección 1/e:", interseccion_1_e_17)
print("Intersección 1/e^2:", interseccion_1_e_c_17)
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def plot_gaussian_fit(x, y, A_fit, mu_fit, sigma_fit, label, y_err=20, uno_e_value=None, uno_e_c_value=None, titulo=None):
    # Encontrar las coordenadas de intersección
    x_interseccion_1_e = fsolve(lambda x: gaussiana(x, A_fit, mu_fit, sigma_fit) - uno_e_value, 0)
    x_interseccion_1_e_c = fsolve(lambda x: gaussiana(x, A_fit, mu_fit, sigma_fit) - uno_e_c_value, 0)

    # Crear la figura y los ejes
    plt.figure(figsize=(10, 6))

    # Scatter plot de los datos con barras de error de ±y_err (en mV)
    plt.errorbar(x, y, yerr=y_err, label='Datos',color='navy', fmt='bo', capsize=5)

    # Gráfico del ajuste exponencial
    x_fit = np.linspace(min(x), 40, 100)  # Rango extendido hasta 40 para el ajuste
    plt.plot(x_fit, gaussiana(x_fit, A_fit, mu_fit, sigma_fit), label='Ajuste Exponencial', color='dodgerblue')

    # Gráfica de las líneas horizontales 1/e y 1/e^2 si se proporcionan los valores
    if uno_e_value is not None:
        plt.axhline(y=uno_e_value, color='k', linestyle='--', label='$1/e$')
    if uno_e_c_value is not None:
        plt.axhline(y=uno_e_c_value, color='purple', linestyle='--', label='$1/e^2$')

    # Personalizar el gráfico
    if titulo is not None:
        plt.title(titulo)
    plt.xlabel('Pasos x 25 $ \mu m$')
    plt.ylabel(' Amplitud [V]')
    plt.legend()

    # Etiquetas de los puntos de intersección
    if uno_e_value is not None:
        plt.annotate(f'Intersección 1/e: ({x_interseccion_1_e[0]:.2f}, {uno_e_value:.2f})',
                     xy=(x_interseccion_1_e[0], uno_e_value), xytext=(x_interseccion_1_e[0] - 10, uno_e_value + 0.1),
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.3"))
    if uno_e_c_value is not None:
        plt.annotate(f'Intersección $1/e^2$: ({x_interseccion_1_e_c[0]:.2f}, {uno_e_c_value:.2f})',
                     xy=(x_interseccion_1_e_c[0], uno_e_c_value),
                     xytext=(x_interseccion_1_e_c[0] - 10, uno_e_c_value + 0.1),
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.3"))

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()

```

```python
cinturaR = 'CinturaR.xlsx'

df = pd.read_excel(cinturaR)
df.head()
```

```python
#df = df.iloc[:16]
df['1e2_micro'] = df['1e2'] * 25/2
df['dist_x'] = df['Corte'] * 25*4*df['vueltas']
df['-1e2_micro']= df['1e2_micro']*(-1)
df
```

```python
from scipy.optimize import curve_fit

# Supongamos que tienes datos en dos arrays: x y y

# Definir la función parabólica que quieres ajustar
def parabola(x, a, b, c):
    return a * x**2 + b * x + c

# Ajustar la función a los datos
params_1, covariance_1 = curve_fit(parabola, df['dist_x'],df['1e2_micro'] )
params_2, covariance_2 = curve_fit(parabola, df['dist_x'],df['-1e2_micro'] )
a_1, b_1, c_1 = params_1
a_2, b_2, c_2 = params_2

# Generar una serie de puntos x para la gráfica
x_fit = np.linspace(-1000, max(df['dist_x']), 1000)
y_fit_1 = parabola(x_fit, a_1, b_1, c_1)
y_fit_2 = parabola(x_fit, a_2, b_2, c_2)

# Grafica el índice (índice del DataFrame) contra 'nueva_columna'



plt.figure(figsize=(9, 4))
plt.plot(x_fit, y_fit_1, color='darkorange')
plt.plot(x_fit, y_fit_2, label='Ajuste Parabólico', color='darkorange')
plt.plot(df['dist_x'], df['1e2_micro'],color='royalblue')
plt.errorbar(df['dist_x'], df['1e2_micro'], yerr=15, color='royalblue', fmt='o', capsize=5)

plt.plot(df['dist_x'], df['-1e2_micro'],  color='royalblue')
plt.errorbar(df['dist_x'], df['-1e2_micro'], yerr=15, label='Radio del Haz [$1/e^2$]',color='royalblue', fmt='o', capsize=5)

plt.axhline(y=0, color='red', linestyle='--', label='Centro del Haz', linewidth=3.5)
plt.xlabel('Distancia x [$\mu m$]')
plt.ylabel('Radio del Haz [$\mu m$]')
plt.title('Gráfico Cintura de Rayleigh')
plt.legend()
plt.grid(True)

# Muestra el gráfico
plt.show()
```
